### Ingest constructors.json file

##### Step 1 - Read the JSON file using the spark dataframe reader

```python
from pandas import read_csv,read_json
from lib import configuration
from lib import common_functions
```

```python
spark = common_functions.get_spark_session()
```

```python
constructors_schema = "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"
```

```python
constructor_df = spark.read \
.schema(constructors_schema) \
.json(f"{configuration.bronze_folder_path}/constructors.json")

constructor_df = read_json(f'{configuration.bronze_folder_path}/constructors.json', lines=True)
display(constructor_df)
```

```python
constructor_dropped_df = constructor_df.drop(columns='url')
constructor_dropped_df["ingestion_date"] = common_functions.get_ingestion_date()
constructor_dropped_df["data_source"] = configuration.v_data_source
constructor_dropped_df
```

```python
constructor_final_df = constructor_dropped_df.rename(columns={"constructorId":"constructor_id","constructorRef":"constructor_ref"})
constructor_final_df
```

```python
constructor_final_df.to_csv(f"{configuration.silver_folder_path}/constructors.csv", index=False)
```
