##### Produce driver standings

```python
from pandas import read_csv
from lib import configuration
```

```python
race_results_df = read_csv(f'{configuration.silver_folder_path}/race_results.csv')
#race_results_df = race_results_df[(race_results_df['race_year'].isin([2020]))]
race_results_df
```

```python
wins_df = race_results_df[(race_results_df['position']=='1')]
wins_df['position'] = (wins_df['position']).astype(int)
wins_df
```

```python
aggs = {
    'position':'sum'
}

wins_df = wins_df.groupby(["race_year", "team"]).agg(aggs).reset_index()
wins_df = wins_df.rename(columns={"position":"wins"})
wins_df.sort_values('wins', ascending=False)
```

```python
aggs = {
    'points':'sum'
}

constructor_standings_df = race_results_df.groupby(["race_year", "team"]).agg(aggs).reset_index()
constructor_standings_df.sort_values('points', ascending=False)
```

```python
# Left Outer Join
constructor_standings_df = constructor_standings_df.set_index(["team","race_year"]).join(wins_df.set_index(["team","race_year"]), rsuffix='_win', how='left').reset_index()
constructor_standings_df = constructor_standings_df[["race_year", "team", "points", "wins"]]
constructor_standings_df.sort_values('points', ascending=False)
```

```python
constructor_standings_df['rank'] = constructor_standings_df.groupby('race_year')['points'].rank(ascending=False)
constructor_standings_df.sort_values('rank', ascending=True)
```

```python
constructor_standings_df.to_csv(f"{configuration.silver_folder_path}/constructor_standings.csv", index=False)
```
